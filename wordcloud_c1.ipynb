# WordCloud
**WordCloud** is a data visualization technique used for **representing text data** in which the **size of each word indicates its frequency** or **importance**. Significant textual data points can be highlighted using a wordcloud. Wordclouds are widely used for analyzing data from social network websites and can be especially useful in sentiment analysis. 

```python
#Read text file
with open ("texts.txt", "r") as f:
    text=f.read()
    print(text)
```

```python
# Preprocess the text: make lowercase and join paragraphs
cleaned_text = text.lower().replace("\n", " ")
print(cleaned_text)
```

 ## import the string library to use string.punctuation. 
This library provides convenient tools and constants for string manipulation, and string.punctuation is a constant that contains all standard punctuation characters.

## Use str.maketrans
- It's an efficient way to handle character replacements or deletions.
- When combined with .translate(), it can process entire strings without needing to loop over them manually.

**Syntax**

**str.maketrans(from_chars, to_chars, delete_chars)**
- **from_chars:** Characters you want to replace.
- **to_chars:** Characters you want to replace them with (must match the length of from_chars).
- **delete_chars:** Characters you want to remove completely.


```python
#import string library
import string

# "" (from_chars): No characters are being replaced.
# "" (to_chars): No replacement characters are provided.
# string.punctuation (delete_chars): All punctuation characters (!"#$%&'()*+,-./:;<=>?@[\\]^_{|}~`) will be deleted.
table = str.maketrans("", "", string.punctuation)

## Use the table to translate the string (removes punctuation)
words = cleaned_text.translate(table)
print(words)
```

```python
# This splits the text into individual words by separating them wherever there's a space. Breaks the text into manageable parts (words) so you can process each word individually.
words = words.split()
print(words)
```

## Convert the List Back to a String

```python
#Use the .join() method to merge the list elements into one string, with spaces between the words.
# Convert the list back to a string
words = " ".join(words)
print(words)
```

```python

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

#store english stopwords in stop_w
stop_w = set(STOPWORDS)
#new_words = ['essential', 'help']
#new_stopwords = stop_w.union(new_words)
print(stop_w)
```

## WordCloud() function has many arguments. To see all arguments of the function, use ?WordCloud

```python
word_cloud = WordCloud(stopwords=stop_w, background_color = "indigo", max_font_size = 50, max_words=100, colormap='rainbow').generate(words)

#Display the wordcloud!
plt.figure(figsize=(10,8))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Give the wordcloud a certain shape e.g following python logo

```python
import numpy as np
import PIL.Image #PIL: Python Imaging Library

#import shape file
pic_pizzahut = PIL.Image.open('pizzahutimage.jpg')

#use numpy array to allow the image take a certain shape
pic_pizzahut = np.array(PIL.Image.open('pizzahutimage.jpg'))

word_cloud = WordCloud(stopwords=stop_w, background_color = "lightgrey", max_font_size = 50, max_words=100, colormap='rainbow', mask=pic_pizzahut).generate(words)

#Display the wordcloud!
plt.figure(figsize=(10,8))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#store to file
word_cloud.to_file("wordcloud_output.png")
```
